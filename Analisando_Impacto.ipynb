# Carregamento de dados

```python
# Custumers - age - valor ausente

media_idade = df_sales_customers['age'].mean()
df_sales_customers['age'] = df_sales_customers['age'].fillna(media_idade)

display(df_sales_customers)
```

```python
# rentals - daily_rate com valor < 0
df_sales_rentals.loc[df_sales_rentals['daily_rate'] <= 0,'daily_rate'] = pd.NA

# Usando valor médio
df_sales_rentals['daily_rate'] = pd.to_numeric(df_sales_rentals['daily_rate'], errors='coerce')
media_daily_rate = df_sales_rentals['daily_rate'].mean()
df_sales_rentals['daily_rate'] = df_sales_rentals['daily_rate'].fillna(media_daily_rate)
display(df_sales_rentals)
```

```python
# venhicles - co2_g_per_km dado ausente

media_co2 = df_sales_vehicles['co2_g_per_km'].mean()
df_sales_vehicles['co2_g_per_km'] = df_sales_vehicles['co2_g_per_km'].fillna(media_co2)
display(df_sales_vehicles)
```

```python
# rentals - daily_rate esta como string
df_sales_rentals['daily_rate'] = pd.to_numeric(df_sales_rentals['daily_rate'], errors='coerce')
display(df_sales_rentals)
```

```python
# rentals - data invalida
# rentals - data com formação diferente
df_sales_rentals['start_date'] = pd.to_datetime(df_sales_rentals['start_date'],errors='coerce')
df_sales_rentals['end_date'] = pd.to_datetime(df_sales_rentals['end_date'],errors='coerce')
display(df_sales_rentals)
```

# Correções

* Calcule a duyração de cada localição (em dias)
* Calcule a receita total por locação
* Faça o merge com a base de veiculos
* Calcule a emissão de CO2 por Locação
* Extraia ano e mês da data de inicio da locação


```python
#Criar backuup de dados
df_sales_carbon_clean = df_sales_carbon
df_sales_rentals_clean = df_sales_rentals
df_sales_customers_clean = df_sales_customers
df_sales_vehicles_clean = df_sales_vehicles
```

```python
#Calcule a duração de cada locação (em dias)

df_sales_rentals_clean['dia_locacao'] = (df_sales_rentals_clean['end_date'] - df_sales_rentals_clean['start_date']).dt.days
display(df_sales_rentals_clean)
```

```python
# Calcule a receita total por locação
# dias de locação * valor da diaria
df_sales_rentals_clean['receita_total'] = df_sales_rentals_clean['dia_locacao'] * df_sales_rentals_clean['daily_rate']
display(df_sales_rentals_clean)
```

```python
# Faça o merge com a base de veículos.
df_consolidado = df_sales_rentals_clean.merge(df_sales_vehicles_clean, on='vehicle_id', how='left')

df_consolidado = df_consolidado.merge(df_sales_customers_clean, on='customer_id', how='left')
display(df_consolidado)
```

```python
# Calcule a emissão de CO₂ por locação.
df_consolidado['CO2_emitido'] = df_consolidado['co2_g_per_km'] * df_consolidado['distance_km']
display(df_consolidado)
```

```python
# Extraia ano e mês da data de início da locação.
#para extraor informações da data temos
# dt.days - dia
# dt.month - mês
# dt.year - ano
# dt.weekday - dia da semana
df_consolidado['mes'] = df_consolidado['start_date'].dt.month
df_consolidado['ano'] = df_consolidado['start_date'].dt.year
df_consolidado['dia_da_semana'] = df_consolidado['start_date'].dt.weekday

display(df_consolidado)
```
